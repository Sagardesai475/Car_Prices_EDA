```python
# drop the unique value
df.drop("vin", axis = 1, inplace=True)
```

```python
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

numerical_columns
```

```python
df = df[numerical_columns]

df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset="sellingprice", inplace=True)
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training (80%) and test (20%) sets
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

y_train = df_train["sellingprice"]
x_train = df_train.drop("sellingprice", axis = 1)

y_test = df_test["sellingprice"]
x_test = df_test.drop("sellingprice", axis = 1)

print(x_train[:3])
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_train
```

```python
from sklearn.impute import SimpleImputer

simple_imputer = SimpleImputer(strategy='mean')

x_train[x_train.columns] = simple_imputer.fit_transform(x_train)
x_test[x_test.columns] = simple_imputer.transform(x_test)
```

```python
x_train.head()
```

```python
x_train.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train[x_train.columns] = scaler.fit_transform(x_train)
x_test[x_test.columns] = scaler.transform(x_test)
```

```python
scaler.feature_names_in_
```

```python
scaler.mean_
```

```python
x_train.head()
```

```python
# Train the Linear Regression model
model = LinearRegression()
model.fit(x_train["condition"].values.reshape(-1,1), y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test["condition"].values.reshape(-1,1))
y_pred
```

```python
#Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# Visualize the true labels vs predicted labels
plt.figure(figsize=(10, 4))

years = scaler.inverse_transform(x_test)[:,1]

sns.lineplot(x=years, y=y_test.values.flatten(), label='True Labels')
sns.lineplot(x=years, y=y_pred.flatten(), label='Predicted Labels')

plt.title('Linear Regression: True Labels vs Predicted Labels')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
